```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Ensure exports directory exists
os.makedirs('../exports', exist_ok=True)

# Set styles
%matplotlib inline
plt.style.use('seaborn-v0_8')  # Updated style name
sns.set_palette('pastel')

# Load data
df = pd.read_csv('../exports/survey_data.csv')

# Display first few rows
print("Data Overview:")
display(df.head())

# Basic statistics
print("\nBasic Statistics:")
display(df.describe())

# Visualization 1: Ages with Highest Income
plt.figure(figsize=(12, 6))
sns.barplot(x='age', y='total_income', data=df, errorbar=None)  # Updated parameter
plt.title('Average Income by Age')
plt.xlabel('Age')
plt.ylabel('Average Income ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../exports/income_by_age.png', dpi=300, bbox_inches='tight')
plt.show()

# Visualization 2: Gender Distribution Across Spending Categories
expense_cols = ['utilities', 'entertainment', 'school_fees', 'shopping', 'healthcare']
melted_df = df.melt(id_vars=['gender'], value_vars=expense_cols, 
                   var_name='category', value_name='amount')

plt.figure(figsize=(14, 8))
sns.boxplot(x='category', y='amount', hue='gender', data=melted_df)
plt.title('Spending Distribution by Gender and Category')
plt.xlabel('Expense Category')
plt.ylabel('Amount Spent ($)')
plt.legend(title='Gender')
plt.tight_layout()
plt.savefig('../exports/spending_by_gender.png', dpi=300, bbox_inches='tight')
plt.show()

# Visualization 3: Healthcare Spending by Age and Gender
plt.figure(figsize=(12, 6))
sns.scatterplot(x='age', y='healthcare', hue='gender', size='total_income',
               sizes=(20, 200), alpha=0.7, data=df)
plt.title('Healthcare Spending by Age and Gender (Bubble Size = Income)')
plt.xlabel('Age')
plt.ylabel('Healthcare Spending ($)')
plt.tight_layout()
plt.savefig('../exports/healthcare_spending.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import os
import pandas as pd
import numpy as np

# Create exports directory if it doesn't exist
os.makedirs('../exports', exist_ok=True)

# Generate sample data if file doesn't exist
if not os.path.exists('../exports/survey_data.csv'):
    # Create sample data
    np.random.seed(42)
    data = {
        'age': np.random.randint(18, 65, 100),
        'gender': np.random.choice(['Male', 'Female'], 100),
        'total_income': np.random.normal(5000, 1500, 100).round(2),
        'utilities': np.random.normal(300, 50, 100).round(2),
        'entertainment': np.random.normal(200, 40, 100).round(2),
        'school_fees': np.random.normal(400, 100, 100).round(2),
        'shopping': np.random.normal(250, 60, 100).round(2),
        'healthcare': np.random.normal(350, 80, 100).round(2)
    }
    
    df = pd.DataFrame(data)
    # Save to CSV
    df.to_csv('../exports/survey_data.csv', index=False)
    print("Sample data created at '../exports/survey_data.csv'")
else:
    print("Data file already exists")
```
